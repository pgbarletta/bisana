```julia
t = Trajectory("/home/pbarletta/labo/20/bisana/run/2lao-1lst/pdbs/2lao.pdb")
p = read(t)
```

```julia
top = Topology(p)
```

```julia
resnames = [ name(Residue(top, i)) for i in 0:convert(Int64, count_residues(top))-1 ]
atm_cnt = Array{Int64, 1}(undef, length(resnames))
for i = 1:length(resnames)
    if resnames[i] == "GLY"
        atm_cnt[i] = 4
    elseif resnames[i] == "ALA"
        atm_cnt[i] = 5
    else
        atm_cnt[i] = 6
    end
end
```
