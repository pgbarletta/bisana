```julia
function write_mtx(output_mtx, file::AbstractString)
    nrows = size(output_mtx)[1]
    ncols = size(output_mtx)[2]
    
    open(file, "w") do f
        for i in 1:nrows
            for j in 1:ncols
                @printf f "%12.4E" output_mtx[i, j]
            end
            @printf f "%s" "\n"
        end
    end
end

function rango_de_archivo(input_mtx, rango::AbstractRange, file::AbstractString)
    mtx = input_mtx[:, rango]
    nrows = size(mtx)[1]
    open(file, "w") do f
        for i in 1:nrows
            for j in rango
                @printf f "%12.4E" mtx[i, j]
            end
            @printf f "%s" "\n"
        end
    end
end
```

### Missings
Los residuos q tienen missings son:

 - GLN 4
 - LYS 110
 - LYS 176
 - GLU 177
 - LYS 188
 - LYS 189
 - LYS 225

A los 8 les falta las cadenas laterales a partir del carbono beta, así q tienen 5 componentes XYZ (15) donde deberían tener 6 XYZ (18). Agarro sus modos y repito los últimos 3 p/ c/u de ellos.

```julia
t = Trajectory(joinpath(home, "pdbs", "2lao.pdb"))
p = read(t)
top = Topology(p)
resnames = [ name(Residue(top, i)) for i in 0:convert(Int64, count_residues(top))-1 ]
atm_cnt = []
faltantes = [ 4, 44, 110, 176, 177, 188, 189, 225 ]
idx_faltantes = []

for i = 1:length(resnames)
    if resnames[i] == "GLY"
        push!(atm_cnt, 4)
    elseif resnames[i] == "ALA"
        push!(atm_cnt, 5)
    else
        push!(atm_cnt, 6)
    end
    if findall(x -> x == i, faltantes) != []
        push!(idx_faltantes, sum(atm_cnt) * 3)
    end
end
sum(atm_cnt)
```

```julia
for i in collect(0:24)
    modos_in = readdlm(joinpath(home, "eigenvectors", string(i, "_modes_2lao")))
    modos_out = Array{Float64}(undef, 0, size(modos_in)[2])
    prev = 1
    for j = 1:length(idx_faltantes)
        x_CB = idx_faltantes[j] - 3
        z_CB = idx_faltantes[j] - 1
        x_R = idx_faltantes[j]
        z_R = idx_faltantes[j] + 2

        modos_out = vcat(modos_out, modos_in[prev:z_CB, :])
        modos_out = vcat(modos_out, modos_in[x_CB:z_CB, :])
        if j == length(idx_faltantes)
            modos_out = vcat(modos_out, modos_in[z_CB+1:end, :])
        end

        global prev = idx_faltantes[j]
    end
    rango_de_archivo(modos_out, 1:10, joinpath(home_10, "eigenvectors", string(i, "_modes_2lao")))
    println(i)
end
```
